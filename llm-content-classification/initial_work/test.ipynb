```python
# how many retweets actually exist for them

client = pymongo.MongoClient("mongodb://localhost:27017/")
user1 = client["rachel_chess_community"]["user_info"]
user2 = pd.DataFrame(list(user1.find()))
ids = set(user2['userid'])


for user in ids:
    og_tweet = client["rachel_chess_community"]["original_tweets"]
    og_tweets_df = pd.DataFrame(list(og_tweet.find({"user_id": user})))
    if len(og_tweets_df) > 0:
        og_ids_set = set(og_tweets_df['id'])
        #print(f"{user}: tweet count:{len(og_ids_set)}")

    retweets = client["rachel_chess_community"]["retweets_of_in_community"]
    retweets_df = pd.DataFrame(list(retweets.find({"retweet_user_id": user})))
    if len(retweets_df) >0:
        retweet_ids_set = set(retweets_df['retweet_id'])
        #print(f"{user}: tweet count:{len(retweet_ids_set)}")

    retweets = client["rachel_chess_community"]["retweets_of_in_community"]
    all_retweets=  pd.DataFrame(list(retweets.find()))
    retweets2 = pd.DataFrame(list(retweets.find({"user_id": user})))
    local_follower = user2[user2["userid"] == user]['local follower list'].to_list()[0]
    
    all_retweets_by_follower = all_retweets[all_retweets['user_id'].isin(local_follower)]
    all_retweets_by_follower.loc[:, 'retweet_id'] = all_retweets_by_follower['retweet_id'].astype(int)
    
    count = 0
    for index, tweet in retweets2.iterrows():
        created_at = tweet['created_at']
        tweet_id = int(tweet['retweet_id'])
        same_retweets = all_retweets_by_follower[all_retweets_by_follower['retweet_id'] == tweet_id]
        same_retweets2 = same_retweets[same_retweets['created_at'] > created_at]
        count += len(same_retweets2)
    print(f"{int(user)}: tweet count:{len(og_ids_set)}, retweet_count:{len(retweet_ids_set)}, last_count: {count}, total: {len(og_ids_set) + len(retweet_ids_set) + count}")

# count = 0
# for tweet in retweet_ids_set:
#     if tweet in og_ids_set:
#         count += 1
```
